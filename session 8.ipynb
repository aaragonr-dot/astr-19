```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace (0, 1, 100)
def get_sin(x):
    return np.sin(x)
def get_cos(x):
    return np.cos(x)
y_sin = get_sin(x)
y_cos = get_cos(x)
```

```python
fig, (ax1, ax2)=plt.subplots(1,2, figsize=(10,4))

ax1.plot(x, y_sin,label='sin(x)', color='purple')
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.grid(True)

ax2.plot(x,y_cos, label='cos(x)', color ='orange')
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.grid(True)
plt.tight_layout()
plt.show()
plt.savefig('figure.pdf',bbox_inches='tight')
```
